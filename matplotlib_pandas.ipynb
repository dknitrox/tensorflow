```python
import matplotlib.pyplot as plt
import numpy as np;

dict = ['yellow','black','blue','red']

a=np.array(range(1,20,5))
print(a)
d=[1,2,3,24]
b=[1,2,24,25]

plt.scatter(x=d,y=b,s=a,c=dict,alpha=0.7)

plt.xscale('log')
plt.xticks([1,5,10,20],['1k','2k','3k','4k'])
plt.text(1,2,'Example')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

paises=["Nueva Zelanda","Peru","España","EE.UU","Japon"]

rangoFechas=pd.date_range(start='1/10/1995',end='1/15/2016',freq='W-MON')

# Elegiremos n paises al azar
np.random.seed(100)
paisesElegidos=[paises[np.random.randint(0,len(paises))] for i in range(0,len(rangoFechas))]
print(paisesElegidos[:5])

numHabitantes=[np.random.randint(100000,1000000) for i in range(0,len(rangoFechas))]
data=list(zip(paisesElegidos,numHabitantes,rangoFechas))
print(data)

# Guardamos la data en un excel

df=pd.DataFrame(data=data,columns=["Pais","Habitantes","Fecha"])
df.to_csv('estudio.csv',index=False,header=True)
```
